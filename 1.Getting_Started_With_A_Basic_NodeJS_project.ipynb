# Setting up a Node.js + Express + MongoDB Backend

## Prerequisites
Make sure you have the following installed:
- [Node.js](https://nodejs.org/)
- [MongoDB](https://www.mongodb.com/)
- A code editor (VS Code recommended)

---

## Step 1: Create a New Project

Open your terminal in the directory where you want to create your project and run:

```sh
mkdir NodeExpressMongoDb
cd NodeExpressMongoDb
```

This creates a new folder and navigates into it.

---

## Step 2: Initialize a Node.js Project

Run the following command to create a `package.json` file:

```sh
npm init -y
```

This generates a `package.json` file that looks like this:

```json
{
  "name": "nodeexpressmongodb",
  "version": "1.0.0",
  "main": "index.js",
  "scripts": {
    "test": "echo \"Error: no test specified\" && exit 1"
  },
  "keywords": [],
  "author": "",
  "license": "ISC",
  "description": ""
}
```

---

## What is `package.json`?

The `package.json` file is a crucial component of any Node.js project. It contains metadata about the project and manages dependencies. Below are the key properties:

- **name**: The name of the project.
- **version**: The current version of the project.
- **main**: The main entry file (e.g., `index.js`).
- **scripts**: Commands that can be run using `npm run <script>`.
- **dependencies**: List of packages required for production (not present initially).
- **devDependencies**: List of packages required for development (not present initially).
- **license**: Specifies the type of license for the project.

## Step 7: Install Express

Express is a fast and minimalist web framework for Node.js. It simplifies server creation and routing.

To install Express, run:

```sh
npm i express
```

After installation, the `node_modules` folder and `package-lock.json` file will be created:
- **node_modules/**: Contains all installed dependencies.
- **package-lock.json**: Ensures consistent dependency versions across environments.

The `dependencies` section in `package.json` will now include:

```json
"dependencies": {
  "express": "^4.21.2"
}
```




Screenshot (9628).png

## Step 8: Modify `index.js`

Replace the existing `index.js` file with the following code:

```js
const express = require('express');
const app = express();

app.listen(3000, () => {
    console.log('Server running on port 3000');
});
```

Now, run the server:

```sh
npm run serve
```

You should see:

```
Server running on port 3000
```

---

## Step 9: Create a Basic Route

Append the following code to `index.js`:

```js
app.get('/', (req, res) => {
    res.send("Hello from Node API");
});
```

### What does `app.get()` do?

- Defines a route handler for HTTP GET requests to `/`.
- When accessed in a browser, it sends a response: `Hello from Node API`.

Now, restart the server:

```sh
ctrl + c  # Stop the server
npm run serve
```

Then, open your browser and go to:

```
http://localhost:3000
```

You should see:

```
Hello from Node API
```

## Step 7: Create a `.gitignore` File

In the project directory, create a `.gitignore` file and add:

```
node_modules
```

### Why use `.gitignore`?
- Prevents tracking unnecessary files in Git.
- `node_modules` contains large dependency files that should be reinstalled via `package.json`.

## Step 8: Install `nodemon` for Development

`nodemon` automatically restarts the server when changes are detected.

To install as a dev dependency, run:

```sh
npm i nodemon -D
```

Modify `package.json` scripts:

```json
"scripts": {
    "test": "echo \"Error: no test specified\" && exit 1",
    "serve": "node index.js",
    "dev": "nodemon index.js"
}
```

Now, run:

```sh
npm run dev
```

This will start the server with `nodemon`, making development easier!
