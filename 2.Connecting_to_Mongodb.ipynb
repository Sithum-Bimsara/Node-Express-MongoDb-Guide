# Connecting to MongoDB Atlas

## Step 1: Sign Up for MongoDB Atlas
1. Go to [MongoDB Atlas](https://www.mongodb.com/atlas/database).
2. Sign up or log in to your account.

## Step 2: Create a New Project
1. Click on `Projects` in the top right.
2. Click `New Project`.

Screenshot (9604).png

3. Enter a project name of your choice and click `Next`.

Screenshot (9605).png

4. On the next page, you can add team members and set permissions (optional). This feature allows multiple developers to collaborate on the database with different permission levels.


Screenshot (9606).png

5. Click `Create Project`.

## Step 3: Create a Cluster
1. Click the `+ Create` button to create a cluster.
Screenshot (9607).png

2. Select the `Free` plan.

Screenshot (9608).png

3. Enter a cluster name (e.g., `BackendDB`).
4. Click `Create Deployment` at the bottom left.


Screenshot (9609).png

## Step 4: Configure Database User
1. After the cluster is created, you will be redirected to a page called "Connect to BackendDB" (or your chosen cluster name).
2. Set a username and password for your database.
3. Save these credentials in a secure place on your local machine.
4. Click `Create Database User`.

Screenshot (9614).png

## Step 5: Choose a Connection Method
1. Click on `Choose a connection method` in below page.


Screenshot (9615).png

2. Select `Drivers` as the connection method.

Screenshot (9616).png

3. On the next page, you will see the Node.js driver configuration.
4. Run the following command in your terminal to install the MongoDB driver:
   ```sh
   npm install mongodb
   ```
5. Copy the connection string provided and save it in a secure place.
6. Click `Done`.

Screenshot (9621).png

## Step 6: Configure Network Access
1. Click on the `Network Access` tab on the right sidebar.

Screenshot (9623).png

2. Click the `Edit` button.

Screenshot (9624).png

3. Select `Allow access from anywhere`.
4. The Access List Entry will be set to `0.0.0.0/0`.
5. Click `Confirm`.

Screenshot (9625).png

## Step 7: Connect MongoDB to Your Node.js Project

1. Stop the nodemon by `Ctrl+C` and run `npm install mongoose` .

2. Change `index.js` file like this,

```
const express = require('express')
const mongoose = require('mongoose')
const app = express()

app.listen(3000, () => {
    console.log('Server running on port 3000')
});

app.get('/', (req, res) => {
    res.send("Hello from Node API")
});

mongoose.connect("mongodb+srv://sithum22:nYaK9Lkc0h8aqtLs@backenddb.nbpav.mongodb.net/?retryWrites=true&w=majority&appName=BackendDB")
.then(() => {
    console.log('Connected to the database!');
})
.catch(() => {
    console.log("Connection failed!");
})

```
